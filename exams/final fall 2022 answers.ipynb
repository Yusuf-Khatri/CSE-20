For example:

If num = 0, sumDigits = 0

If num = 314, sumDigits = 3 + 1 + 4 = 8

Hint: the sum of digits of an integer n is equal to n % 10 + the sum of digits in n // 10

```python
# Your code goes here

# This should work
assert sumDigits(0) == 0
assert sumDigits(314) == 8
```

```python
# Answer
def sumDigits(num):         # 0.5 point for correctly define function
  if num == 0:              # 0.5 point for considering num = 0 and return 0
    return 0
  else :                          
    last_digit = num % 10   # 1 point for finding each digit using modulo of 10
    num = num // 10         # 1 point for setting the new number using floor division
    sum = last_digit + sumDigits(num)  # 1 point for calculating sum of digits using recursion
    return sum

    # Or
    # return num % 10 + sumDigits(num // 10)

print(sumDigits(314))
```
